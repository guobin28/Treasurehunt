```python

from flask import Flask, request
from twilio.twiml.messaging_response import MessagingResponse
import re

app = Flask(__name__)

# 🧠 Player progress (use DB later for real deployment)
players = {}

# 🎮 Game levels
game = [
    {
        "answers": ["apple"],
        "clue": "🍏 Level 2: I have keys but open no locks. What am I?"
    },
    {
        "answers": ["keyboard"],
        "clue": "⌚ Level 3: I have hands but cannot clap. What am I?"
    },
    {
        "answers": ["clock"],
        "clue": "🎉 You finished the game! Well done!"
    }
]

# 🧼 Normalize function (removes spaces, punctuation, case issues)
def normalize(text):
    return re.sub(r"[^a-z0-9]", "", text.lower())


@app.route("/webhook", methods=["POST"])
def webhook():
    incoming_msg = request.form.get("Body", "")
    user = request.form.get("From")

    resp = MessagingResponse()
    msg = resp.message()

    # Normalize user input
    user_msg = normalize(incoming_msg)

    # 🆕 New player
    if user not in players:
        players[user] = 0
        msg.body("🎮 Welcome to the Treasure Hunt!\n\n🍏 Level 1: A fruit that keeps the doctor away?")
        return str(resp)

    level = players[user]

    # 🏁 Game finished check
    if level >= len(game):
        msg.body("🏆 You already completed the game!")
        return str(resp)

    correct_answers = game[level]["answers"]

    # ✅ Check answer (multiple allowed)
    if any(user_msg == normalize(a) for a in correct_answers):
        players[user] += 1
        new_level = players[user]

        # 🎉 Next level or finish
        if new_level < len(game):
            msg.body("✅ Correct!\n\n" + game[new_level]["clue"])
        else:
            msg.body("🏆 Congratulations! You finished all levels!")

    else:
        msg.body("❌ Wrong answer. Try again!")

    return str(resp)


# 🚀 Run server (IMPORTANT for Render + local testing)
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True)


    
```
